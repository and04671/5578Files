```python
# The following script takes input from the user and displays
# it in a frame based on user-defined formatting
#
# SKILLS:
#       1.decision making with if-elif-else
#       2.handling user input
#
# Author: COLE ANDERSON
#
# Score: 10/10
# Comments: Nothing in the repo


# USER INPUT -----------------------------------------------------
sentence = input("Write a sentence here: ")
print("-------------------------------------")
print("Format your sentence -> use the keys:")
print("<c> Capitalize")
print("<u> Upper Case")
print("<l> Lower Case")
print("<r> replace x with y")
print("-------------------------------------")
# *********** PUT YOUR CODE HERE *********************************
# Ask the user to select the format type (c, u, or l) and store 
# the input in a variable named format (we can't use that, is reserved word)
# USER INPUT OF FORMAT TYPE --------------------------------------
# ****************************************************************
desiredFormat = input("What type of format do you want? ") # replace with actual user input
# The following line of code is using a string function
# What kind of reformating does the lower function do?
#Good logic here
desiredFormat = desiredFormat.lower() # reformats the user input
#print(desiredFormat)
# *********** PUT YOUR CODE HERE *********************************
# Check format type selected by the user (letter)
# Make sure that upper case and lower case letters are accepted. 
# yup, that's what the initial lower is for
# if 'c' then use .capitalize to capitalize
if desiredFormat == "c":
    sentence = sentence.capitalize()
    #print(sentence)
# if 'u' then use an .upper to put to upper case        
elif desiredFormat == "u":
    sentence = sentence.upper()
    #print(sentence)
# if 'l' then use .lower to put to lower case    
elif desiredFormat == "l":
    sentence = sentence.lower()
    #print(sentence) 
# if 'r' use another input to replace characters as desired
elif desiredFormat == "r":
    #Nice flexiblity
    old = input("Ok, what do you want to replace? ")
    new = input("and...what do you want to replace that with? ")
    old = old.lower()
    sentence = sentence.replace(old, new)
    #print(sentence)
# Finally, if the user accidentally typed something else
# then inform the user about the mistake and put the sentence
# to some default (e.g. do no user formatting)
else:
    sentence = sentence
    #print(sentence)
# ****************************************************************
# FRAME PARAMETERS -----------------------------------------------
# I think we're supposed to build a frame here, so I'm gonna do that...
sentenceLength = len(sentence)
#print(sentenceLength)
#bounds the sentence in a frame that changes based on text length
print("$" + "-" * (sentenceLength + 8) + "$")
print("|    " + sentence + "    |")
print("$" + "-" * (sentenceLength + 8) + "$")

# SENTENCE STATISTICS --------------------------------------------
print("sentence length:", sentenceLength)
print("Checking for occurences of letters <a>, <b>, and <c>")
sentence = sentence.lower() # Why do that? Explain briefly!
# We need to do that to make all the letters of one type readable as the same. Otherwise A will not 
# register the same as a. 

# *********** PUT YOUR CODE HERE *********************************
# Use the proper string function to check how many
# times 'a' occurs in sentence, do the same for 'b' and 'c'
# ****************************************************************
a_in_sentence = sentence.count("a")# determine the correct value of a's using .count
if a_in_sentence == 0:
   print("There is no <a> in: " + sentence)
else:
   print("<a> occured", a_in_sentence, "times")

# *********** PUT YOUR CODE HERE *********************************
# Write equivalent code that checks the number of 'b' and 'c'
# ****************************************************************
b_in_sentence = sentence.count("b")# determine the correct value of b's using .count
if b_in_sentence == 0:
   print("There is no <b> in: " + sentence)
else:
   print("<b> occured", b_in_sentence, "times")

c_in_sentence = sentence.count("c")# determine the correct value of c's using .count
if c_in_sentence == 0:
   print("There is no <c> in: " + sentence)
else:
   print("<c> occured", c_in_sentence, "times")
```
