```python
# The following script calculates a number of statistics based
# on a predefined 2-dimensional array (grid)
# SKILLS:
#       1.looping
#       2.the 'list' object
#       3.built-in functions
#       4.module importing
# Author: Cole Anderson
# 
# Score: 10/10
# Comments: Code is good. Nice and clean.
# INPUT MATRIX------------------
grid = [[1,1,2,4,1,7,1,7,6,9],\
        [1,2,5,3,9,1,1,1,9,1],\
        [7,4,5,1,8,1,2,0,0,4],\
        [1,4,1,1,1,1,1,1,8,5],\
        [9,0,0,0,0,0,1,1,9,8],\
        [7,4,2,1,8,2,2,2,9,7],\
        [7,4,2,1,7,1,1,1,0,5],\
        [3,4,5,3,4,5,9,1,0,9],\
        [0,0,5,1,1,1,9,7,7,7]]

print('')
print("Our grid *********************")

for i in grid:
   print(i)
        
# SIZE OF MATRIX ---------------
# row num
rows = len(grid)
# col num
cols = len(grid[0])
# cell num
celNum = rows * cols

print("\nNumber of rows", rows)
print("\nNumber of columns", cols)
print("\nNumber of cells", celNum)
print("\n")
#stats display-----------------------------
# ****** PUT YOUR CODE HERE **********
# Hint: the easiest way to calculate
# the stats is to convert the
# 2D grid into a 1D array (list)
# ************************************
gridList = []
for i in grid:
    for j in i:
        gridList.append(j)
#print(gridList)
# ****** PUT YOUR CODE HERE **********
# CALCULATE AND print(SUM ------------
print("The sum of values is:", sum(gridList))
# CALCULATE AND print(AVERAGE --------
print("The average is:", sum(gridList)/len(gridList))
# CALCULATE AND print(MIN ------------
print("The smallest value is:", min(gridList))
# CALCULATE AND print(MAX ------------
print("The largest value is:", max(gridList))
# CALCULATE AND print(RANGE ----------
print("The range of values is:", max(gridList) - min(gridList))
# ************************************
```
